```python
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.make_forge()
X_train, X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train,y_train)
```

```python
print("Test set predictions: ", clf.predict(X_test))
```

```python
#Checking accuracy of the model
print("Test set accuracy: {:.2f}".format(clf.score(X_test,y_test)))
```

```python
fig, axes = plt.subplots(1,3,figsize=(10,3))

for n_neighbors, ax in zip([1,3,9], axes):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(knn,X,fill=True,eps=0.5,ax=ax)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title("{} lneighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3) 
```

```python
from sklearn.datasets import load_breast_cancer
cancer_df = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer_df.data, cancer_df.target, stratify=cancer_df.target, random_state=66)
training_accuracy=[]
test_accuracy=[]
#setting neighbors from 1 to 10 
neighbors_settings = range(1,11)
for n_neighbors in neighbors_settings:
    knn1 = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn1.fit(X_train,y_train)
    training_accuracy.append(knn1.score(X_train,y_train))
    test_accuracy.append(knn1.score(X_test,y_test))
plt.plot(neighbors_settings,training_accuracy, label="training accuracy")
plt.plot(neighbors_settings,test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```
